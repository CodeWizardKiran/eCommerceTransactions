```python
# Importing libraries
import pandas as pd

# Loading the datasets
customers = pd.read_csv('customers.csv')
products = pd.read_csv('products.csv')
transactions = pd.read_csv('transactions.csv')
```

```python
print("\nTransactions dataset preview:")
print(transactions.head())
print(transactions.tail())
transactions.info()
transactions.describe()
```

```python
# Checking for missing values before handling them
print("Missing values before filling:")
print("\nCustomers dataset:")
print(customers.isnull().sum())  # Display missing values for 'customers' dataset
```

```python
print("\nProducts dataset:")
print(products.isnull().sum())  # Display missing values for 'products' dataset
```

```python
print("\nTransactions dataset:")
print(transactions.isnull().sum())  # Display missing values for 'transactions' dataset
```

```python
# Removing duplicates if any
customers = customers.drop_duplicates()  # Dropping duplicates in 'customers' dataset
products = products.drop_duplicates()  # Dropping duplicates in 'products' dataset
transactions = transactions.drop_duplicates()  # Dropping duplicates in 'transactions' dataset
```

```python
# Handling missing values using forward fill method (ffill)
customers = customers.ffill()  # Forward fill missing values in 'customers' dataset 
products = products.ffill()  # Forward fill missing values in 'products' dataset
transactions = transactions.ffill()  # Forward fill missing values in 'transactions' dataset
# there are No mission values but I just mentioned how to handle them with ffill() method
```

```python
# Verifying if any missing values remain after filling
print("\nMissing values after handling:")
print("\nCustomers dataset:")
print(customers.isnull().sum())  # Check if any missing values remain in 'customers' dataset

print("\nProducts dataset:")
print(products.isnull().sum())  # Check if any missing values remain in 'products' dataset

print("\nTransactions dataset:")
print(transactions.isnull().sum())  # Check if any missing values remain in 'transactions' dataset
```

```python
#Customer Segmentation
customer_segmentation = transactions.groupby('CustomerID').agg(
    TotalSales=('TotalValue', 'sum'),
    TransactionCount=('TransactionID', 'count')
).reset_index()
customer_segmentation = customer_segmentation.sort_values(by='TotalSales', ascending=False)

# Visualize top 10 customers
plt.figure(figsize=(12, 6))
sns.barplot(data=customer_segmentation.head(10), x='CustomerID', y='TotalSales', palette='coolwarm')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)
plt.show()
```

```python
#Impact of Pricing on Sales
price_analysis = transactions.groupby('Price').agg(
    TotalSales=('TotalValue', 'sum'),
    QuantitySold=('Quantity', 'sum')
).reset_index()

# Scatter plot of price vs quantity sold
plt.figure(figsize=(12, 6))
sns.scatterplot(data=price_analysis, x='Price', y='QuantitySold', color='orange')
plt.title('Impact of Price on Quantity Sold')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.grid()
plt.show()
```

```python
# Inventory Visualisation
inventory_analysis = transactions.groupby('ProductID').agg(
    TotalSales=('TotalValue', 'sum'),
    QuantitySold=('Quantity', 'sum'),
    AveragePrice=('Price', 'mean')
).reset_index()

# Example visualization: Top products by sales and stock
plt.figure(figsize=(12, 6))
sns.barplot(data=inventory_analysis.sort_values(by='QuantitySold', ascending=False).head(10),
            x='ProductID', y='QuantitySold', palette='magma')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product ID')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```
